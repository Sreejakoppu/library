```python
import pandas as pd
```

```python
df = pd.read_csv("MELBOURNE_HOUSE_PRICES_LESS.csv" )
```

```python
df.head(2)
```

```python
df.tail(2)
```

```python
#no.of rows and cols
print(df.shape)
```

```python
#drop a problematic data
df=df.dropna()
```

```python
print(df.shape)
```

```python
rooms=df["Rooms"]
print(rooms)
```

```python
price=list(df["Price"])
```

```python
import matplotlib.pyplot as plt
```

```python
fig=plt.figure()
plt.hist(x=price,color="green",edgecolor="black",bins=20)
plt.xlabel("Price range")
plt.ylabel("number of houses")
plt.title("Histogram of Price")
plt.show()
```

```python
rooms_count=df["Rooms"].value_counts()
```

```python
print(rooms_count)
```

```python
#bar plot
#type
type_count=df["Type"].value_counts()
type_count=dict(type_count)
print(type_count)
```

```python
k=list(type_count.keys())
v=list(type_count.values())
```

```python
fig=plt.figure(figsize=(4,4))
plt.bar(x=k,height=v)
plt.show()
```

```python
#pie chart
rooms_count=df["Rooms"].value_counts()
rooms_count=dict(rooms_count)
print(rooms_count)
```

```python
a=rooms_count.keys()
b=rooms_count.values()
```

```python
print(a,b)
```

```python
#pie
fig=plt.figure(figsize=(7,7))
plt.pie(x=b,labels=a)
plt.legend()
plt.title("Pie chart")
plt.show()
```

```python
regionname_count=df["Regionname"].value_counts()
regionname_count=dict(regionname_count)
r=regionname_count.keys()
rc=regionname_count.values()
fig=plt.figure(figsize=(8,8))
plt.title("Pie chart")
plt.pie(x=rc,labels=r)
plt.legend()
plt.show()
```

```python
#scatter plot
price=df["Price"]
price=list(price)
distance=df["Distance"]
distance=list(distance)
```

```python
fig=plt.figure(figsize=(7,7))
plt.plot(distance,price,linestyle="",marker="o",markerfacecolor="pink",markeredgecolor="blue")
plt.xlabel("Distance")
plt.ylabel("Price")
plt.title("Distance vs Price")
plt.show()
```
